```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings("ignore")

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
main_df = pd.read_csv('insurance.csv')
```

```python
main_df
```

```python
X = main_df.drop(columns=['charges'])
y = main_df['charges']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = pd.get_dummies(X, drop_first=True, dtype='int')
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=4)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
linreg = LinearRegression()
```

```python
linreg.fit(X_train, y_train)
```

```python
y_preds_test = linreg.predict(X_test)
y_preds_train = linreg.predict(X_train)
```

```python
sns.scatterplot(x=y_test, y=y_preds_test)
```

```python
sns.scatterplot(x=y_train, y=y_preds_train, hue = X_train['region_northwest'])
```

```python
np.random.seed(0)
X = 2 * np.random.rand(100, 1)  # Independent variable
y = 4 + 3 * X + np.random.randn(100, 1)  # Dependent variable with noise

# Train the linear regression model
model = LinearRegression()
model.fit(X, y)

# Make predictions
X_new = np.array([[0], [2]])  # New data points for prediction
y_pred = model.predict(X_new)

# Plot the data points
plt.scatter(X, y, color='blue', label='Sample Data')

# Plot the regression line
plt.plot(X_new, y_pred, color='red', linewidth=2, label='Linear Regression')

# Customize plot
plt.xlabel('X')
plt.ylabel('y')
plt.title('Simple Linear Regression')
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```

```python
# Generate some sample data
np.random.seed(0)
X = 2 * np.random.rand(25, 1)  # Independent variable
y = 4 + 3 * X + np.random.randn(25, 1)  # Dependent variable with noise

# Train the linear regression model
model = LinearRegression()
model.fit(X, y)

# Make predictions
X_new = np.array([[0], [2]])  # New data points for prediction
y_pred = model.predict(X_new)

# Plot the data points
plt.scatter(X, y, color='blue', label='Sample Data')

# Plot the regression line
plt.plot(X_new, y_pred, color='red', linewidth=2, label='Linear Regression')

# Customize plot
plt.xlabel('X')
plt.ylabel('y')
plt.title('Simple Linear Regression with Vertical Lines')
plt.legend()
plt.grid(True)

plt.savefig("regression1.png")
# Plot vertical lines connecting each point to the regression line
for i in range(len(X)):
    y_i = model.predict(X[i].reshape(1, -1))[0]  # Extracting the scalar value from the prediction array
    plt.plot([X[i], X[i]], [y[i], y_i], color='green', linestyle='--', linewidth=1)

plt.savefig("regression2.png")
# Show plot
plt.show()
```
